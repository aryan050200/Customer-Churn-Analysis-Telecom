```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 1: Importing Libraries and Loading Dataset

In this step, required Python libraries are imported and the customer churn dataset
is loaded into a Pandas DataFrame for analysis.


```python
df = pd.read_csv('Customer Churn.csv')
df.head()
```

```python
df.info()
```

#replacing blanks with 0 as tenure is 0 and no total charges are recorded 

## Step 2: Data Cleaning

During data inspection, it was observed that the `TotalCharges` column
contained blank spaces for customers with zero tenure.
These values were replaced with 0 and converted to float type
to ensure numerical analysis can be performed.


```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

#converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Churn', data=df)

ax.bar_label(ax.containers[0])
plt.title("Customer Churn Distribution in Telecom Services")
plt.xlabel("Churn Status")
plt.ylabel("Number of Customers")

plt.show()
```

The visualization shows that a larger proportion of customers have not churned,
indicating overall stable customer retention.


```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customeres", fontsize = 10)
plt.show()
```

The pie chart indicates that approximately 26.54% of customers have churned.
This highlights a significant customer loss and emphasizes the need to further
analyze the factors contributing to churn.


```python
plt.figure(figsize = (3,3))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Plot
fig, ax = plt.subplots(figsize=(4, 4))  # Adjust figsize for better visualization

# Plot the bars
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])  # Customize colors if desired

# Add percentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

plt.title('Churn by Senior Citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', bbox_to_anchor = (0.9,0.9))  # Customize legend location

plt.show()
```

Comparatively, a higher percentage of customers in the senior citizen category
have churned, indicating that age may be an important factor influencing churn behavior.


```python
plt.figure(figsize = (9,4))
sns.histplot(x = "tenure", data = df, bins = 72, hue = "Churn")
plt.show()
```

Customers with longer service tenure tend to remain with the company,
while customers who have used the services for only 1–2 months
show a higher likelihood of churning.


```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "Contract", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Contract")
plt.show()
```

Customers on month-to-month contracts show higher churn rates,
suggesting long-term contracts help retain customers.


```python
df.columns.values
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of columns for the subplot grid (you can change this)
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate number of rows needed

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))  # Adjust figsize as needed

# Flatten the axes array for easy iteration (handles both 1D and 2D arrays)
axes = axes.flatten()

# Iterate over columns and plot count plots
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Remove empty subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

The analysis shows that customers who do not churn are more likely to have
services such as Phone Service, Internet Service (particularly DSL),
and Online Security enabled.
In contrast, higher churn rates are observed among customers who do not use
or do not have access to services like Online Backup, Tech Support,
and Streaming TV.
 

```python
plt.figure(figsize=(8, 4))
ax = sns.countplot(x="PaymentMethod", data=df, hue="Churn")

ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])

plt.title("Customer Churn by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)

plt.show()
```

A higher churn rate is observed among customers who pay via electronic check,
suggesting that payment method may influence customer retention behavior.


## Conclusion

This analysis highlights that customer churn is strongly influenced by
contract type, tenure, and payment method.
Customers with short-term contracts and lower tenure are more likely to churn.
The findings suggest that offering long-term plans and improving early customer
engagement can help reduce churn rates.

## 📌 Key Business Insights

- Customers on **month-to-month contracts** show significantly higher churn compared to long-term contracts.
- Customers with **low tenure (less than 12 months)** are more likely to churn.
- Users paying via **Electronic Check** have higher churn rates than those using automatic payment methods.
- Customers with **higher monthly charges** tend to churn more frequently.


## 💡 Business Recommendations

- Introduce **discounts or loyalty offers** for month-to-month customers to encourage long-term commitment.
- Promote **annual and two-year contracts** with price benefits.
- Encourage customers to switch to **automatic payment methods** by offering small incentives.
- Focus retention campaigns on **new customers (low tenure)** through onboarding support and offers.


## 🏁 Final Conclusion

This customer churn analysis identifies key factors influencing customer attrition in the telecom domain.
By understanding customer behavior related to contract type, tenure, and payment methods, businesses can
